```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
P=50 #poplulation
F=1 #constant (Scaling factor)
CR = 0.55 # crossover ratio
```

```python
A= np.random.rand(1,50)*2*np.pi

A=A[0].tolist()
A[:7]
```

```python
pre_best=0
i=0
```

# Mutation

```python
#Mutation
def Mutation():
    M=[]
    for i in range(P):
        r =(np.random.rand(3)*49).astype(int)
        r1=r[0]
        r2=r[1]
        r3=r[2]
        while(r1==i or r2==i or r3==i):
            r =(np.random.rand(3)*49).astype(int)
            r1=r[0]
            r2=r[1]
            r3=r[2]
        
        M.append(A[r1] +F*(A[r2]-A[r3]))

        
    return M
```

# Cross over

```python
def cross_over(M,A):
    C=[]
    for i  in range(P):
        r =np.random.rand(1)
        if r[0] <=CR:
            C.append(M[i])
        else:
            C.append(A[i])
    return C
        
```

# function 

```python
def function(value):
    return np.sin(value)
```

# Fitness function 

```python
def fitness_fun(C,A):
    X=[]
    for i in range(P):
        if(function(C[i]>function(A[i]))):
            X.append(C[i])
        else:
            X.append(A[i])
    return X
```

# checking in range 

```python
def check_range(X):
    for i in range(P):
        if(X[i]>np.pi*2 or X[i]<0):
            temp=np.random.rand(1)*2*np.pi;
            X[i]=temp[0]
    return X
```

# plot

```python
def plot(A):
    global i
    i=i+1
    y=np.linspace(0,2*np.pi)
    plt.plot(y,np.sin(y))
    plt.xlabel('Genratioin {}'.format(i))
    plt.scatter(A,np.sin(A))
```

```python
plot(A)
```

```python


while 1 :
    plot(A)
    M =Mutation()
    C =cross_over(M,A)
    X =fitness_fun(C,A)
    X =check_range(X)
    Y =[np.sin(i) for i in X  ]
    X_list= list(zip(X,Y))
    X_list.sort(key=lambda t :t[1])
    A=[]
    A=[X_list[i][0] for  i in  np.linspace(0,49,dtype=int)]
    best =np.sin(A[49])
    if  math.fabs(pre_best-best)<0.0000000000000001   and i>20:
        break;
    else:
   
        pre_best=best
    # Wait for 2 seconds
    plt.show()
  
    



```
